# Numpypy

Numpypy es numpy para pypy. 

¿Por qué crear numpy para pypy?

* Se encuentra entre una de las peticiones que más han demandado los usuarios de pypy.
* Pretenden convertir a Python (pypy en este caso) en la herramienta principal para realizar cálculo numérico en lugar de que Python sea un lenguaje pegamento (glue language) para otras librerías escritas en C/C++, fortran,...

Más información en la página de donación para [Numpy dentro de pypy](http://pypy.org/numpydonate.html).

Extraido del status oficial que podéis encontrar [aquí](http://buildbot.pypy.org/numpy-status/latest.html)

```python
%%pypy
import numpy as np
from numpy import sin, pi
import time

def plate_displacement(xx, yy, ww, a, b, P, xi, eta, D, max_m, max_n):
    max_i, max_j = ww.shape
    for mm in range(1, max_m):
        for nn in range(1, max_n):
            for ii in range(max_i):
                for jj in range(max_j):
                    a_mn = 4*P * sin(mm * pi * xi / a) * sin(nn * pi * eta / b) / (a * b)
                    ww[ii, jj] += (a_mn / (mm**2 / a**2 + nn**2 / b**2)**2
                                   * sin(mm * pi * xx[ii, jj] / a)
                                   * sin(nn * pi * yy[ii, jj] / b)
                                   / (pi**4 * D))

# Plate geometry
a = 1.0  # m
b = 1.0  # m
h = 50e-3  # m

# Material properties
E = 69e9  # Pa
nu = 0.35

# Series terms
max_m = 16
max_n = 16

# Computation points
# NOTE: With an odd number of points the center of the place is included in
# the grid
NUM_POINTS = 101

# Load
P = 10e3  # N
xi = a / 2
eta = a / 2

# Flexural rigidity
D = h**3 * E / (12 * (1 - nu**2))

# ---

# Set up domain
x = np.linspace(0, a, num=NUM_POINTS)
y = np.linspace(0, b, num=NUM_POINTS)
xx, yy = np.meshgrid(x, y)

# Compute displacement field
ww = np.zeros_like(xx)
plate_displacement(xx, yy, ww, a, b, P, xi, eta, D, max_m, max_n)

# Print maximum displacement
w_max = abs(ww).max()
print("Maximum displacement = %14.12f mm" % (w_max * 1e3))
print("alpha = %7.5f" % (w_max / (P * a**2 / D)))
print("alpha * P a^2 / D = %6.4f mm" % (0.01160 * P * a**2 / D * 1e3))

t = []
times = 10
for i in range(times):
    t0 = time.time()
    plate_displacement(xx, yy, ww, a, b, P, xi, eta, D, max_m, max_n)
    t1 = time.time() - t0
    t.append(t1)
print(sum(t) / times)
print(t)
```
